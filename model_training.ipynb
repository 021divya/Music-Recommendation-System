```python
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

def token(text):
    tokens = word_tokenize(text)
    stemmed = [stemmer.stem(word) for word in tokens]
    return " ".join(stemmed)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
Tfid=TfidfVectorizer(analyzer='word',stop_words='english')
```

```python
matrix = Tfid.fit_transform(df['text'])
```

```python
similar=cosine_similarity(matrix)
```

```python
def recommender(song_name):
    # Normalize and match song name (case-insensitive and trimmed)
    idx_match = df[df['song'].str.strip().str.lower() == song_name.strip().lower()].index

    # Check if song was found
    if idx_match.empty:
        return ["Song not found in dataset."]
    
    idx = idx_match[0]

    distance = sorted(
        list(enumerate(similar[idx])),
        reverse=True,
        key=lambda x: x[1]
    )

    song = []
    for s_id in distance[1:6]:  # Skip first one as it is the song itself
        song.append(df.iloc[s_id[0]]['song'])

    return song
```

```python
recommender("Start A New Life")
```

```python
recommender("America, The Beautiful")
```
